### Funciones de obtención de matrices COM

```python
# COM horizaontal
def sp_COMMaker_h(imagen, marker, label_n, norm = False): 
    matriz_auxiliar = np.zeros((np.amax(imagen)+1,)*2, dtype=int)
    dimensiones = imagen.shape
    normalizador = 0
    for i in range(dimensiones[0]):
        for j in range(dimensiones[1]-1):
            if (marker[i][j] == label_n and marker[i][j+1] == label_n):
                matriz_auxiliar[imagen[i][j]][imagen[i][j+1]] += 1
                normalizador += 2
    COM = matriz_auxiliar + np.transpose(matriz_auxiliar)
    return (COM/normalizador) if norm else COM

# COM vertical
def sp_COMMaker_v(imagen, marker, label_n, norm = False): 
    matriz_auxiliar = np.zeros((np.amax(imagen)+1,)*2, dtype=int)
    dimensiones = imagen.shape
    normalizador = 0
    for i in range(dimensiones[0]-1):
        for j in range(dimensiones[1]):
            if (marker[i][j] == label_n and marker[i+1][j] == label_n):
                matriz_auxiliar[imagen[i][j]][imagen[i+1][j]] += 1
                normalizador += 2
    COM = matriz_auxiliar + np.transpose(matriz_auxiliar)
    return (COM/normalizador) if norm else COM

# COM diagonal (esq.inf.izq -> esq.sup.der.)
def sp_COMMaker_45(imagen, marker, label_n, norm = False): 
    matriz_auxiliar = np.zeros((np.amax(imagen)+1,)*2, dtype=int)
    dimensiones = imagen.shape
    normalizador = 0
    for i in range(1,dimensiones[0]):
        for j in range(dimensiones[1]-1):
            if (marker[i][j] == label_n and marker[i-1][j+1] == label_n):
                matriz_auxiliar[imagen[i][j]][imagen[i-1][j+1]] += 1
                normalizador += 2
    COM = matriz_auxiliar + np.transpose(matriz_auxiliar)
    return (COM/normalizador) if norm else COM

# COM diagonal (esq.sup.izq. -> esq.inf.der.)
def sp_COMMaker_135(imagen, marker, label_n, norm = False): 
    matriz_auxiliar = np.zeros((np.amax(imagen)+1,)*2, dtype=int)
    dimensiones = imagen.shape
    normalizador = 0
    for i in range(1,dimensiones[0]):
        for j in range(1,dimensiones[1]):
            if (marker[i][j] == label_n and marker[i-1][j-1] == label_n):
                matriz_auxiliar[imagen[i][j]][imagen[i-1][j-1]] += 1
                normalizador += 2
    COM = matriz_auxiliar + np.transpose(matriz_auxiliar)
    return (COM/normalizador) if norm else COM
```

### Funciones de obtención de características

```python
def entropy(COM_n):
    entropia = 0
    for renglon in COM_n:
        for value in renglon:
            entropia += (value*log(value)) if (value>0) else 0
    return -entropia

def homogeneity(COM_n):
    homogeneidad = 0
    Y, X = COM_n.shape
    for i in range(Y):
        for j in range(X):
            homogeneidad += COM_n[i][j]/(1+abs(i-j))
    return homogeneidad

#Angular Second Moment, Energy
def smoothness(COM_n):
    ASM = 0
    Y, X = COM_n.shape
    for i in range(Y):
        for j in range(X):
            ASM += COM_n[i][j]**2
    return ASM

def contrast(COM_n,k=2,n=1):
    contraste = 0
    Y, X = COM_n.shape
    for i in range(Y):
        for j in range(X):
            contraste += ((i-j)**k)*(COM_n[i][j]**n)
    return contraste
```

```python
def sp_COMs(imagen, marker):
    vcs_COM = []
    for nSP in np.unique(marker):
        COM_h = sp_COMMaker_h(rgb2gray(imagen), marker, nSP, True)
        #COM_v = sp_COMMaker_v(rgb2gray(imagen), marker, nSP, True)
        vcs_COM.append([entropy(COM_h), smoothness(COM_h)])
    return vcs_COM
```

### Funciones Auxiliares

```python
def rgb2gray(imagen):
    gris = ((imagen[:,:,0]+imagen[:,:,1]+imagen[:,:,2])/3).astype(int)
    return gris
```

```python
def mostrarSP(superpixeles, clasi, colores, imagen, guardar='aux'):
    superpixel_aux = np.zeros(imagen.shape, dtype=np.uint8)
    Y,X = superpixeles.shape
    proporciones = []
    aux = 0
    for i in range(Y):
        for j in range(X):
            aux = clasi[superpixeles[i,j]-1]
            proporciones.append(aux)
            superpixel_aux[i,j] = colores[aux]
    plt.imshow(superpixel_aux)
    plt.axis('off')
    plt.show()
    pltI.imsave('./resultados/Kmeans'+guardar+'.png', superpixel_aux)
    print('Clase0: ',proporciones.count(0))
    print('Clase1: ',proporciones.count(1))
    print('Clase2: ',proporciones.count(2))
```

```python
def get_SP(imagen, nombre='', **kwargs):
    show = (kwargs["show"] if ("show" in kwargs) else False)
    save = (kwargs["save"] if ("save" in kwargs) else False)
    i_sp = slic(imagen, n_segments=600, start_label=1, compactness=8.0, sigma=1.0)
    if show:
        plt.imshow(mark_boundaries(imagen, i_sp))
        plt.axis('off')
        plt.title(nombre, color='darkgray')
        plt.show()
    if save:
        pltI.imsave('sp/sp'+nombre+'_'+str(np.amax(i_sp))+'_n600_c8_s1'+nombre[11:21]+'.png', mark_boundaries(imagen, i_sp))
    return i_sp
```

```python
def get_SP_hsv(imagen, nombre='', **kwargs):
    show = (kwargs["show"] if ("show" in kwargs) else False)
    save = (kwargs["save"] if ("save" in kwargs) else False)
    i_sp = slic(imagen, n_segments=600, start_label=1, compactness=8.0, sigma=1.0)
    if show:
        plt.imshow(mark_boundaries(imagen, i_sp))
        plt.axis('off')
        plt.title(nombre, color='darkgray')
        plt.show()
    if save:
        pltI.imsave('sp/sp_'+nombre+'HSV_'+str(np.amax(i_sp))+'_n600_c8_s1'+nombre[11:21]+'.png', mark_boundaries(imagen, i_sp))
    return i_sp
```
